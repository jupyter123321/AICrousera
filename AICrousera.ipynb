```python
import pandas as pd
import numpy as np
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
```

```python
# define regression model
def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data.head()
```

```python
concrete_data.shape
```

```python
concrete_data.describe()
```

```python
concrete_data.isnull().sum()
```

```python
concrete_data_columns = concrete_data.columns

predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
target = concrete_data['Strength'] # Strength column
```

```python
predictors.head()
```

```python
target.head()
```

```python
n_cols = predictors.shape[1] # number of predictors
```

```python
model = regression_model()
```

```python
#model.fit(predictors_norm, target, validation_split=0.3, epochs=50, verbose=0)
list_of_mean_squared_error = []
for cycle in range(50):
    #Randomly split the data into a training set (70%) and a test set (30%):  
    x_train, x_test, y_train, y_test = train_test_split(predictors, target, test_size=0.3)
    #Train and test the model at the same time
    res = model.fit(x_train, y_train, epochs=50, verbose=0, validation_data=(x_test, y_test))
    #Find mean_squared_error as last value in history.
    mean_squared_error = res.history['val_loss'][-1]
    #Add value of mean_squared_error for every cycle in list.
    list_of_mean_squared_error.append(mean_squared_error)
    print('Repeat Number: {}: MSE: {}'.format(cycle+1, mean_squared_error))
```

```python
print('Mean for all squared errors: {}'.format(np.mean(list_of_mean_squared_error)))
print('STD for mean squared errors: {}'.format(np.std(list_of_mean_squared_error)))
```

```python
#Task Number 2 (B)
def regression_model2():
    model2 = Sequential()
    model2.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model2.add(Dense(1))
    
    model2.compile(optimizer='adam', loss='mean_squared_error')
    return model2

model2 = regression_model2()
```

```python
predictors_norm = (predictors - predictors.mean()) / predictors.std()
predictors_norm.head()
```

```python
n_cols = predictors_norm.shape[1] # number of predictors
```

```python
#model.fit(predictors_norm, target, validation_split=0.3, epochs=50, verbose=0)
list_of_mean_squared_error_norm = []
for cycle in range(50):
    #Randomly split the data into a training set (70%) and a test set (30%):  
    x_train, x_test, y_train, y_test = train_test_split(predictors_norm, target, test_size=0.3)
    #Train and test the model at the same time
    res = model2.fit(x_train, y_train, epochs=50, verbose=0, validation_data=(x_test, y_test))
    #Find mean_squared_error as last value in history.
    mean_squared_error_norm = res.history['val_loss'][-1]
    #Add value of mean_squared_error for every cycle in list.
    list_of_mean_squared_error_norm.append(mean_squared_error_norm)
    print('Repeat Number Normalized: {}: MSE: {}'.format(cycle+1, mean_squared_error_norm))
```

```python
print('Mean for all squared errors: {}'.format(np.mean(list_of_mean_squared_error_norm)))
print('STD for mean squared errors: {}'.format(np.std(list_of_mean_squared_error_norm)))
```
